```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Given list of numbers
numbers = [10, 17, 23, 42, 59, 84, 97]

# Use filter to get prime numbers
prime_numbers = list(filter(is_prime, numbers))

print("Prime numbers:", prime_numbers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("C:/Users/Punith Harshitha/Desktop/mtcars.csv")

# Plot the histogram for the 'mpg' column
plt.hist(df['mpg'], bins=10, edgecolor='black')
plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Define the number of bins
bins = 10

# Calculate histogram values
counts, bin_edges = np.histogram(df['mpg'], bins=bins)

# Find the interval with the highest frequency
max_count_index = np.argmax(counts)
highest_interval = (bin_edges[max_count_index], bin_edges[max_count_index + 1])

print(f"Interval with the highest frequency: {highest_interval}")
print(f"Frequency of this interval: {counts[max_count_index]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/Punith Harshitha/Desktop/mtcars.csv")

# Map transmission type for better readability
df['transmission'] = df['am'].map({0: 'Automatic', 1: 'Manual'})

# Count the frequency of each transmission type
transmission_counts = df['transmission'].value_counts()

# Plot the bar plot
plt.figure(figsize=(8, 5))
transmission_counts.plot(kind='bar', color=['blue', 'orange'], edgecolor='black')
plt.title('Frequency Distribution of Transmission Types')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Rotate x-axis labels to horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/Punith Harshitha/Desktop/mtcars.csv")

# Create a box plot for the 'mpg' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['mpg'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'), 
            whiskerprops=dict(color='black'), 
            capprops=dict(color='black'), 
            medianprops=dict(color='red'))
plt.title('Box Plot of MPG')
plt.xlabel('Miles Per Gallon (MPG)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Calculate the five-number summary
min_val = df['mpg'].min()
q1 = df['mpg'].quantile(0.25)
median = df['mpg'].median()
q3 = df['mpg'].quantile(0.75)
max_val = df['mpg'].max()

# Print the five-number summary
print(f"Minimum: {min_val}")
print(f"First Quartile (Q1): {q1}")
print(f"Median (Q2): {median}")
print(f"Third Quartile (Q3): {q3}")
print(f"Maximum: {max_val}")
```
